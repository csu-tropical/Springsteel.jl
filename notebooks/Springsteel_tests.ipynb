```julia
using Springsteel
```

```julia
gp = GridParameters(
        geometry = "XYZ",
        xmin = -10.0,
        xmax = 10.0,
        num_cells = 20,
        ymin = -10.0,
        ymax = 10.0,
        BCL = Dict(
            "u" => CubicBSpline.PERIODIC),
        BCR = Dict(
            "u" => CubicBSpline.PERIODIC),
        BCU = Dict(
            "u" => CubicBSpline.PERIODIC),
        BCD = Dict(
            "u" => CubicBSpline.PERIODIC),
        zmin = 0.0,
        zmax = 10.0,
        zDim = 32,
        BCB = Dict(
            "u" => Chebyshev.R0),
        BCT = Dict(
            "u" => Chebyshev.R0),
        vars = Dict(
            "u" => 1))
```

```julia
grid = createGrid(gp)
```

```julia
calcTileSizes(grid, 1)
```
